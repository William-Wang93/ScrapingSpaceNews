```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
page = requests.get("https://spacenews.com/section/commercial/")
content = page.content
parser = BeautifulSoup(content, 'html.parser')
```

```python
body = parser.body
```

```python
p = body.find_all("p")
p[0].get_text()
```

```python
count = 0
list =[]
for i in p:
    print(p[count].get_text())
    list.append(p[count].get_text())
    count+=1
    
```

```python
df = pd.DataFrame (list,columns=["Summary"])
df.drop(df.tail(2).index,inplace=True)
df
```

```python
h = body.find_all("h2", class_='launch-title')
```

```python
count = 0
listh =[]
for i in h:
    print(h[count].get_text())
    listh.append(h[count].get_text())
    count+=1
```

```python
df['Title']=listh
df
```

```python
list_link=[]
link = body.find_all('h2', class_='launch-title')
countz=0
for i in link:
    list_link.append(link[countz].a.get('href'))
    countz+=1
df['link']=list_link
```

```python
df = df[["Title", "Summary", "link"]]
df
```
