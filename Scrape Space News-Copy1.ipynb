```python
#import libraries
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
#This step is important; it 
page = requests.get("https://spacenews.com/section/launch-section/")
content = page.content
parser = BeautifulSoup(content, 'html.parser')
```

```python
#select Body from paragraph
body = parser.body
```

```python
#Find Paragraph
p = body.find_all("p")
p[0].get_text()
p
```

```python
#This collects all the summaries of the articles, and appends them into a list. 
count = 0
list =[]
for i in p:
    #print(p[count].get_text())
    list.append(p[count].get_text())
    count+=1
print (list)    
```

```python
df = pd.DataFrame (list,columns=["Summary"])
df.drop(df.tail(2).index,inplace=True)
df
```

```python
h = body.find_all("h2", class_='launch-title')
```

```python
#This returns the titles in a list, which can be used appended in a Pandas Datframe
count = 0
listh =[]
for i in h:
    print(h[count].get_text())
    listh.append(h[count].get_text())
    count+=1
```

```python
df['Title']=listh
df
```

```python
#Extract links from Beautiful Soup, 
list_link=[]
link = body.find_all('h2', class_='launch-title')
countz=0
for i in link:
    list_link.append(link[countz].a.get('href'))
    countz+=1
df['link']=list_link
```

```python
#Rearrange columns so the order is Title, Summary, Link
df = df[["Title", "Summary", "link"]]
df
```
